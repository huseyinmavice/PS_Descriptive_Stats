```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
auto_data = pd.read_csv('../datasets/auto_data_pro.csv')
auto_data.sample(5)
```

```python
auto_data['mpg'].plot.kde(figsize=(12,8))

plt.xlabel('mpg')

plt.title('Density plot for MPG')

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.boxplot(x='cylinders', y='mpg', data=auto_data)

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.violinplot(x='cylinders', y='mpg', data=auto_data, innder=None)
sns.swarmplot(x='cylinders', y='mpg', data=auto_data, color='w')

plt.show()
```

```python
cylinder_stats = auto_data.groupby(['cylinders'])['mpg'].agg(['mean', 'count', 'std'])
cylinder_stats
```

```python
ci95_high = []
ci95_low = []
```

```python
for i in cylinder_stats.index:
    mean, count, std = cylinder_stats.loc[i]
    
    ci95_high.append(mean + 1.96 * (std / math.sqrt(count)))
    ci95_low.append(mean - 1.96 * (std / math.sqrt(count)))
```

```python
cylinder_stats['ci95_HIGH'] = ci95_high
cylinder_stats['ci95_LOW'] = ci95_low

cylinder_stats
```

```python
cylinders = 4
cylinders4_df = auto_data.loc[auto_data['cylinders'] == cylinders]
cylinders4_df.sample(10)
```

```python
plt.figure(figsize=(12,8))

sns.distplot(cylinders4_df['mpg'], rug=True, kde=True, hist=False)


plt.stem([cylinder_stats.loc[cylinders]['mean']],
         [0.07], linefmt = 'C1',
         markerfmt = 'C1', label = 'mean')

plt.stem([cylinder_stats.loc[cylinders]['ci95_LOW']],
         [0.07], linefmt = 'C2',
         markerfmt = 'C2', label = '95% CI LOW')

plt.stem([cylinder_stats.loc[cylinders]['ci95_HIGH']],
         [0.07], linefmt = 'C3',
         markerfmt = 'C3', label = '95% CI HIGH')

plt.xlabel('mpg')
plt.legend()
plt.show()
```
